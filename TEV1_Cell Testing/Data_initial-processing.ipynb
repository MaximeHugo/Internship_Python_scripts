```python
import shutil
from pathlib import Path
import pandas as pd
import os
import tkinter as tk
from tkinter import filedialog
from tkinter import messagebox
from tkinter import simpledialog
from tkinter import ttk
```

```python
def copy_subdirectories_and_files(source_root, destination_folder):
    # Create destination folder if it does not exist yet
    Path(destination_folder).mkdir(parents=True, exist_ok=True)
    
    # Go through all subfolders in root directory
    for racine, dossiers, _ in os.walk(source_root):
        for dossier in dossiers:
            # Create the path to source subfolder
            source_folder = os.path.join(racine, dossier)
            # Define destination path to subfolder
            destination_subfolder = os.path.join(destination_folder, dossier)
            # Check first if the path to subfolder already exist 
            if os.path.exists(destination_subfolder):
                for file in os.listdir(source_folder):
                    # Create a copy of all files for each subfolder in the new destination
                    source_file = os.path.join(source_folder, file)
                    destination_file = os.path.join(destination_subfolder, file)

                    if os.path.isfile(source_file): # Check is the object is a file

                        if not os.path.exists(destination_file):

                            try:
                                shutil.copy2(source_file, destination_file) # Copy the file in the new directory
                                print(f"Copied file: {source_file} -> {destination_file}")
                        
                            except Exception as e:
                                print(f"An error occured during file copy {source_file} : {e}")

                        else:
                            print(f"File already exists: {destination_file}, skipping.")

            else:
                # Subfolder copy from root to destination folder
                try:
                    shutil.copytree(source_folder, destination_subfolder)
                    print(f"Copied folder: {source_folder} -> {destination_subfolder}")

                except Exception as e:
                    print(f"An error occured during folder copy {source_folder} : {e}")

source_root = 'BCS 915_172.16.148.45/Raw-Export'  # path adress of source folder
destination_folder = 'BCS 915_172.16.148.45/Raw-Export_Wo_headers'  # path adress of folder to copy the root's subfolder

copy_subdirectories_and_files(source_root, destination_folder)
```

```python
def delete_metadata_lines(files_path, sequence):
    # Read all lines of the file
    for root, folders, files in os.walk(files_path):
        for file_name in files:
            files_dir = os.path.join(root, file_name)

            try:

                with open(files_dir, 'r') as f: # Open the file in read mode
                    lines = f.readlines()

    # Search the line index when the test data appears for the first time using sequence as  condition
                index_sequence = next((i for i, line in enumerate(lines) if sequence in line), None)

    # Open the file in write mode to save in a new file
                with open(files_dir, 'w') as f:
                    if index_sequence is not None:
            # Write the lines starting at the index found
                        f.writelines(lines[index_sequence:])
                    else:
                        print(f"sequence not found in {file_name}")
            except Exception as e:
                print(f"Error during processing of {file_name}: {e}")

sequence = 'Time /s,Step,Task,Cycle number,Loop 1 number,Loop 2 number,Loop 3 number,Loop 4 number,U /V,I /A'  # Define of the string sequence to find in the file
files_path = Path('BCS 915_172.16.148.45/Raw-Export_Wo_headers')   # Definition of the root path

delete_metadata_lines(files_path, sequence)
```

```python
#Select initial folder you want to process data files from, Will be found in Raw-Export_Wo_headers directory 
data_path = Path('BCS 915_172.16.148.45/Raw-Export_Wo_headers')
tk.Tk().withdraw() # prevents an empty tkinter window from appearing
folder_path = Path(filedialog.askdirectory(initialdir = data_path))

file_list = [file for file in folder_path.iterdir() if file.is_file()]

#Select folder in which you want to store the processed data files
output_Path = Path('BCS 915_172.16.148.45/Process-Export')
tk.Tk().withdraw() # prevents an empty tkinter window from appearing
output_directory = Path(filedialog.askdirectory(initialdir = output_Path))
output_directory.mkdir(parents=True, exist_ok=True)

SubFolder_list = ['1C_rate_Cycles', 'C10_rate_Cycles', 'GITT_HPPC_Cycles', 'Graphs_Output']
SubFolder_dict = {}

for SubFolder in SubFolder_list:
    SubFolder_Path = os.path.join(output_directory, SubFolder)

    try:
        os.makedirs(SubFolder_Path, exist_ok = True)
        SubFolder_dict[SubFolder] = SubFolder_Path

    except Exception as e:
        print(f'An issue occured during folder creation: {e}')

for key, value in SubFolder_dict.items():
    print(f'Folder {key} created: {value}')
```

```python
# Extract data for cycles (CC-CV / CC) @ 1C  
param1_1C = 'CC_CV'
param2_1C = 'CC'

# steps to remove for 1C due to current values overlapping
param1_to_Del = 'task_18'
param2_to_Del = 'task_35'
    
# Extract data for cycles @ C/10
param1_C10 = 'CC_CV'
param2_C10 = 'CC'
paramTask_C10 = ['task_18', 'task_20', 'task_35', 'task_37']

# Extract data for GITT and HPPC
param1_GITT_HPPC = 'AUP' #Steps for GITT cycle & HPPC pulses repetitions
param2_GITT_HPPC = 'CC_CV'
param3_GITT_HPPC = 'CC'

param1 = 'task_26'
param2 = 'task_43'
    # First HPPC
param3 = 'task_22'
param4 = 'task_25'
    # Second HPPC
param5 = 'task_39'
param6 = 'task_42'

for file in file_list:
    try:
        # Generate dataframe
        df_list = pd.read_csv(file, low_memory=False)

        # Filter the dataframe for GITT cycle
        df_filter_GITT = df_list[(df_list['Task'].isin([param1_GITT_HPPC])) &
                                 (~df_list['Step'].isin([param1, param2]))]
        # Filter the dataframe for HPPC cycles
        df_filter_HPPC = df_list[(df_list['Step'].isin([param1, param2, param3, param4, param5, param6])) & 
                                 (df_list['Task'].isin([param1_GITT_HPPC, param2_GITT_HPPC, param3_GITT_HPPC])) &
                                 (~df_list.index.isin(df_filter_GITT.index))]
        # Filter the dataframe for cycles at C/10
        df_filter_C10 = df_list[(df_list['Task'].isin([param1_C10, param2_C10])) &
                                (df_list['Step'].isin(paramTask_C10)) &
                                (~df_list.index.isin(df_filter_GITT.index))&
                                (~df_list.index.isin(df_filter_HPPC.index))]
        # Filter the datframe for cycles at 1C
        df_filter_1C = df_list[(df_list['Task'].isin([param1_1C, param2_1C])) &
                               ((df_list['I /A'].between(4.5, 52.5)) | (df_list['I /A'].between(-52.5, -47.5))) &
                               (~df_list.index.isin(df_filter_C10.index)) &
                               (~df_list.index.isin(df_filter_GITT.index))&
                               (~df_list.index.isin(df_filter_HPPC.index))]
        
        base_name = os.path.basename(file)  # Take the base file name
        name_without_extension, _ = os.path.splitext(base_name)  # Separate file name and extension

        # Create new file for each type of cycles (filtered dataframe)
        output_file_1C = os.path.join(SubFolder_dict['1C_rate_Cycles'], f'1C_{name_without_extension}.csv')
        #print(f"File for 1C: {Path(output_file_1C).stem}")
               
        output_file_C10 = os.path.join(SubFolder_dict['C10_rate_Cycles'], f'C10_{name_without_extension}.csv')
        #print(f"File for 0.1C: {Path(output_file_C10).stem}")

        output_file_GITT = os.path.join(SubFolder_dict['GITT_HPPC_Cycles'], f'GITT_{name_without_extension}.csv')
        #print(f"File for GITT: {Path(output_file_GITT).stem}")

        output_file_HPPC = os.path.join(SubFolder_dict['GITT_HPPC_Cycles'], f'HPPC_{name_without_extension}.csv')
        #print(f'File for HPPC: {Path(output_file_HPPC).stem}\n')

        # Save the data in corresponding file
        df_filter_1C.to_csv(output_file_1C, index=False)
        if not df_filter_C10.empty:
            df_filter_C10.to_csv(output_file_C10, index=False)
        if not df_filter_GITT.empty:
            df_filter_GITT.to_csv(output_file_GITT, index=False)
        if not df_filter_HPPC.empty:
            df_filter_HPPC.to_csv(output_file_HPPC, index=False)

        if df_filter_1C.empty or df_filter_C10.empty or df_filter_GITT.empty or df_filter_HPPC.empty:
            print(f'No data found for {Path(output_file_1C).stem}\n'
                  f'No data found for {Path(output_file_C10).stem}\n'
                  f'No data found for {Path(output_file_GITT).stem}\n'
                  f'No data found for {Path(output_file_HPPC).stem}\n')
        else:
            print(f'Data extracted and saved in {Path(output_file_1C).stem}\n'
                  f'Data extracted and saved in {Path(output_file_C10).stem}\n'
                  f'Data extracted and saved in {Path(output_file_1C).stem}\n'
                  f'Data extracted and saved in {Path(output_file_1C).stem}\n')  

    except Exception as e:
        print(f"Error during file processing {file.name}: {e}")
```
